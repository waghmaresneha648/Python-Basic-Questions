## Q1. BMI Calculator

```python
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))
bmi = weight / (height ** 2)
print("Your BMI is:", round(bmi, 2))
```

## Q2. Discount Calculator

```python
original_price = float(input("Enter the original price: "))
discount = float(input("Enter the discount percentage: "))
discounted_price = original_price - (original_price * discount / 100)
print("Discounted price is:", discounted_price)
```

## Q3. Speed Calculator

```python
distance = float(input("Enter the distance in km: "))
time = float(input("Enter the time in hours: "))
speed = distance / time
print("Average speed is:", round(speed, 2), "km/h")
```

## Q4. Username Generator

```python
full_name = input("Enter your full name: ")
birth_year = input("Enter your birth year: ")
first_name = full_name.split()[0]
username = first_name.lower() + birth_year
print("Your username is:", username)
```

## Q5. Palindrome Checker

```python
word = input("Enter a word: ").lower()
if word == word[::-1]:
    print("It's a palindrome!")
else:
    print("Not a palindrome.")
```

## Q6. Word Counter

## Q7. Shopping Cart

```python
cart = ["milk", "bread", "eggs", "butter"]
cart.append("cheese")
cart.remove("bread")
cart.sort()
print("Final shopping cart:", cart)
```

## Q8. Find the Maximum

```python
numbers = [45, 12, 78, 34, 90, 23]
maximum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
print("The largest number is:", maximum)
```

## Q9. Reverse a List

```python
numbers = [10, 20, 30, 40, 50]
print("Original list:", numbers)
print("Reversed list:", end=" ")
for i in range(len(numbers) - 1, -1, -1):
    print(numbers[i], end=" ")
print()
```

## Q10. Phone Book

```python
phonebook = {
    "alice": "9876543210",
    "bob": "8765432109",
    "charlie": "7654321098"
}
name = input("Enter a name to search: ").lower()
if name in phonebook:
    print("Phone number:", phonebook[name])
else:
    print("Contact not found.")
```

## Q11. Inventory Manager

```python
inventory = {"pens": 50, "notebooks": 30, "erasers": 100}
print("Current inventory:", inventory)
item = input("Enter item name: ").lower()
quantity = int(input("Enter quantity to add: "))
if item in inventory:
    inventory[item] += quantity
else:
    inventory[item] = quantity
print("Updated inventory:", inventory)
```

## Q12. Grade Calculator

```python
students = {"Alice": 92, "Bob": 78, "Charlie": 65, "Diana": 55, "Eve": 88}
for name, marks in students.items():
    if marks >= 90:
        grade = "A"
    elif marks >= 75:
        grade = "B"
    elif marks >= 60:
        grade = "C"
    else:
        grade = "F"
    print(f"{name}: {marks} -> Grade {grade}")
```

## Q13. Leap Year Checker

```python
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a Leap Year!")
else:
    print(year, "is not a Leap Year.")
```

## Q14. Triangle Type

```python
a = float(input("Enter side 1: "))
b = float(input("Enter side 2: "))
c = float(input("Enter side 3: "))
if a == b == c:
    print("Equilateral Triangle")
elif a == b or b == c or a == c:
    print("Isosceles Triangle")
else:
    print("Scalene Triangle")
```

## Q15. Voting Eligibility

## Q16. Multiplication Table

## Q17. Sum of Digits

```python
number = input("Enter a number: ")
total = 0
for digit in number:
    total += int(digit)
print("Sum of digits:", total)
```

## Q18. FizzBuzz

```python
for i in range(1, 51):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```
