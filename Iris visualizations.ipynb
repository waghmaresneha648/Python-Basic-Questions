```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
 

iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)
 
print("Dataset shape:", iris.shape)
print(iris.head())
print("\nSpecies count:\n", iris['species'].value_counts())
 

plt.rcParams['figure.figsize'] = (8, 5)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
 
colors_map = {'setosa': 'steelblue', 'versicolor': 'coral', 'virginica': 'mediumseagreen'}
 
 
# ── 1. Bar Chart: Number of Flowers per Species
plt.figure()
plt.title("Number of Flowers per Species", fontsize=14)
counts = iris['species'].value_counts()
plt.bar(counts.index, counts.values,
        color=['steelblue', 'coral', 'mediumseagreen'])
plt.xticks(fontsize=12, rotation=45)
plt.yticks(ticks=np.arange(0, 60, 10), fontsize=12)
plt.ylabel("Count")
plt.tight_layout()
plt.show()
 
 
# ── 2. Adjusting xticks and yticks
plt.figure()
plt.title("Number of Flowers per Species", fontsize=14)
plt.bar(counts.index, counts.values,
        color=['steelblue', 'coral', 'mediumseagreen'])
plt.xticks(fontsize=12, rotation=45)
plt.yticks(ticks=np.arange(0, 60, 10), fontsize=12)
plt.ylabel("Count")
plt.tight_layout()
plt.show()
 
 
# ── 3. Changing default figure size
plt.rcParams['figure.figsize'] = (9, 6)
plt.rc('xtick', labelsize=13)
plt.rc('ytick', labelsize=13)
 
plt.figure()
plt.title("Number of Flowers per Species (Larger Figure)", fontsize=14)
plt.bar(counts.index, counts.values,
        color=['steelblue', 'coral', 'mediumseagreen'])
plt.xticks(rotation=45)
plt.yticks(ticks=np.arange(0, 60, 10))
plt.ylabel("Count")
plt.tight_layout()
plt.show()
 
# Reset back
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
# ── 4. Simple Histogram: Sepal Length
plt.figure()
plt.title("Sepal Length Distribution", fontsize=14)
plt.hist(x=iris['sepal length (cm)'], bins=10, color='steelblue', edgecolor='white')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
 
 
# ── 5. Customizing the Histogram 
plt.figure()
plt.title("Sepal Length Distribution (4cm - 8cm)", fontsize=14)
plt.hist(x=iris['sepal length (cm)'], bins=12, range=(4, 8),
         color='darkgreen', edgecolor='white')
plt.xticks(rotation=45)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# ── 6. Pie Chart: Species Distribution 
plt.figure()
plt.title("Species Distribution", fontsize=14)
plt.pie(counts.values,
        labels=counts.index,
        colors=['steelblue', 'coral', 'mediumseagreen'],
        autopct='%1.1f%%',
        explode=(0.05, 0, 0),
        startangle=90)
plt.tight_layout()
plt.show()
 
 
# ── 7. Histogram: Petal Length Distribution 
plt.figure()
plt.title("Petal Length Distribution (2cm - 7cm)", fontsize=14)
plt.hist(x=iris['petal length (cm)'], bins=12, range=(2, 7),
         color='darkblue', edgecolor='white')
plt.xticks(rotation=45)
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# ── 8. Overlapping Histogram: Sepal Length by Species 
plt.figure()
plt.title("Sepal Length by Species", fontsize=14)
for species, color in colors_map.items():
    subset = iris[iris['species'] == species]
    plt.hist(subset['sepal length (cm)'], bins=10, alpha=0.6,
             color=color, label=species, edgecolor='white')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
 
 
# ── 9. Overlapping Histogram: Petal Width by Species 
plt.figure()
plt.title("Petal Width by Species", fontsize=14)
for species, color in colors_map.items():
    subset = iris[iris['species'] == species]
    plt.hist(subset['petal width (cm)'], bins=10, alpha=0.6,
             color=color, label=species, edgecolor='white')
plt.xlabel("Petal Width (cm)")
plt.ylabel("Count")
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# ── 10. Scatter Plot: Sepal Length vs Sepal Width 
plt.figure()
plt.title("Sepal Length vs Sepal Width", fontsize=14)
for species, color in colors_map.items():
    subset = iris[iris['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'],
                label=species, color=color, alpha=0.7, s=60, edgecolors='white')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
 
 
# ── 11. Scatter Plot: Petal Length vs Petal Width 
plt.figure()
plt.title("Petal Length vs Petal Width", fontsize=14)
for species, color in colors_map.items():
    subset = iris[iris['species'] == species]
    plt.scatter(subset['petal length (cm)'], subset['petal width (cm)'],
                label=species, color=color, alpha=0.7, s=60, edgecolors='white')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
 
```

```python
# ── 12. Box Plot: Sepal Length by Species 
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_title("Sepal Length by Species", fontsize=14)
data_to_plot = [iris[iris['species'] == s]['sepal length (cm)']
                for s in iris_data.target_names]
bp = ax.boxplot(data_to_plot, patch_artist=True,
                labels=iris_data.target_names)
box_colors = ['steelblue', 'coral', 'mediumseagreen']
for patch, color in zip(bp['boxes'], box_colors):
    patch.set_facecolor(color)
ax.set_ylabel("Sepal Length (cm)")
plt.tight_layout()
plt.show()
 
 
# ── 13. Box Plot: Petal Length by Species 
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_title("Petal Length by Species", fontsize=14)
data_to_plot = [iris[iris['species'] == s]['petal length (cm)']
                for s in iris_data.target_names]
bp = ax.boxplot(data_to_plot, patch_artist=True,
                labels=iris_data.target_names)
for patch, color in zip(bp['boxes'], box_colors):
    patch.set_facecolor(color)
ax.set_ylabel("Petal Length (cm)")
plt.tight_layout()
plt.show()
 
 
```

```python
# ── 14. Bar Chart: Average Feature Values per Species 
avg = iris.groupby('species')[iris_data.feature_names].mean()
x = np.arange(len(iris_data.feature_names))
width = 0.25
 
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("Average Feature Values per Species", fontsize=14)
for i, (species, color) in enumerate(colors_map.items()):
    ax.bar(x + i * width, avg.loc[species], width,
           label=species, color=color, edgecolor='white')
ax.set_xticks(x + width)
ax.set_xticklabels(['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'],
                   fontsize=11, rotation=15)
ax.set_ylabel("cm")
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()
```
