```python
#Ask the user for a sentence. Count how many vowels (a, e, i, o, u) are in that sentence.
sentence = input("Enter a sentence: ")
vowels = "aeiou"
vowel_count = 0
for letter in sentence.lower():
    if letter in vowels:
        vowel_count = vowel_count + 1
print("Total number of vowels:", vowel_count)
```

```python
balance = 1000

print(f"Welcome! Your current balance is: ${balance}")

# Ask the user for their choice
action = input("Would you like to 'withdraw' or 'deposit'? ").lower()

if action == "withdraw":
    amount = float(input("Enter the amount to withdraw: "))

    # Check if there are sufficient funds
    if amount <= balance:
        balance -= amount
        print(f"Transaction successful. You withdrew ${amount}.")
    else:
        print("Insufficient funds! You cannot withdraw more than your balance.")

elif action == "deposit":
    amount = float(input("Enter the amount to deposit: "))
    balance += amount
    print(f"Transaction successful. You deposited ${amount}.")

else:
    print("Invalid option selected.")

# Print final balance
print(f"Your final balance is: ${balance}")
```
